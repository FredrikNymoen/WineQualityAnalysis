```python
import sys
from pathlib import Path
import pandas as pd

ROOT = Path.cwd().parents[1] 
sys.path.append(str(ROOT / "src"))

df = pd.read_csv("../../data/wineQT.csv")
```

```python
from utils import plotSeries

# correlation with quality
corr = df.corr(numeric_only=True)["quality"].drop("quality").drop("Id").sort_values(key=abs, ascending=False)



print(corr.sort_values(key=abs, ascending=False))

plotSeries(
    corr,
    kind="bar",
    title="Correlation of Features with Wine Quality",
    xlabel="Features",
    ylabel="Correlation with Quality",
    rotation=45,
    color="darkorange",
    figsize=(10, 5),
    save_path="../figures/featureCorrelation.png"
)
```

Based on the correlation values, the ranking of features from best to worst predictor of wine quality is:
1. Alcohol
2. Volatile acidity
3. Sulphates
4. Citric acid
5. Total sulfur dioxide
6. Density
7. Chlorides
8. Fixed acidity
9. Free sulfur dioxide
10. pH
11. Residual sugar
