```python
import sys
from pathlib import Path
import pandas as pd

ROOT = Path.cwd().parents[1] 
sys.path.append(str(ROOT / "src"))

df = pd.read_csv("../../data/wineQT.csv")
```

```python
from gdLinear import GDLinearReg
from utils import plotSeries
import matplotlib.pyplot as plt

# Q1.3.1: quality ~ chlorides
X_chlorides = df[["chlorides"]].values
y = df["quality"].values

model_chlorides = GDLinearReg(lr=0.001, n_iter=5000)
model_chlorides.fit(X_chlorides, y)
y_pred_chlorides = model_chlorides.predict(X_chlorides)

# plot regression line
plt.scatter(X_chlorides, y, alpha=0.5, label="Data")
plt.plot(X_chlorides, y_pred_chlorides, color="red", label="Regression line")
plt.xlabel("Chlorides")
plt.ylabel("Quality")
plt.title("Simple Linear Regression: Quality ~ Chlorides")
plt.legend()
plt.savefig("../figures/regressionChlorides.png", bbox_inches="tight")
plt.show()
```

```python
# Q1.3.2: quality ~ alcohol
X_alcohol = df[["alcohol"]].values

model_alcohol = GDLinearReg(lr=0.0001, n_iter=5000)
model_alcohol.fit(X_alcohol, y)
y_pred_alcohol = model_alcohol.predict(X_alcohol)

# plot regression line
plt.scatter(X_alcohol, y, alpha=0.5, label="Data")
plt.plot(X_alcohol, y_pred_alcohol, color="red", label="Regression line")
plt.xlabel("Alcohol")
plt.ylabel("Quality")
plt.title("Simple Linear Regression: Quality ~ Alcohol")
plt.legend()
plt.savefig("../figures/regressionAlcohol.png", bbox_inches="tight")
plt.show()
```

```python
print("=== Chlorides model ===")
print("Intercept:", model_chlorides.intercept_) # Where the line crosses the y-axis
print("Coefficient:", model_chlorides.coef_) # Slope of the line

print("\n=== Alcohol model ===")
print("Intercept:", model_alcohol.intercept_) # Where the line crosses the y-axis
print("Coefficient:", model_alcohol.coef_) # Slope of the line
```

```python
# the regression line chlorides is ca flat, meaning that chlorides has little effect on quality.
# the regression line alcohol has a positive slope, meaning that higher alcohol content increases quality.
```
